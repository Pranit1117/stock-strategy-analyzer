```python
import pandas as pd

data = pd.read_csv("../data/nifty50.csv", header=[0,1])
```

```python
data.head()
```

```python

data.columns = data.columns.get_level_values(0)
```

```python
data.head()
```

```python

data = data.iloc[1:]
```

```python
data
```

```python

data["Price"] = pd.to_datetime(data["Price"])
data = data.set_index("Price")
```

```python
data
```

```python
data
```

```python
data.index.name = "Date"
```

```python
data
```

```python
data["returns"] = data["Close"].pct_change()
```

```python
data["volatility_20"] = data["returns"].rolling(20).std()
```

```python
data["sma_20"] = data["Close"].rolling(20).mean()
data["sma_50"] = data["Close"].rolling(50).mean()

data["trend_strength"] = (data["sma_20"] - data["sma_50"]) / data["sma_50"]
```

```python
data["daily_range"] = (data["High"] - data["Low"]) / data["Close"]
data["range_20"] = data["daily_range"].rolling(20).mean()
```

```python
feature_data = data[
    ["returns", "volatility_20", "trend_strength", "range_20"]
].dropna()
```
